```python
from nocode_robot_programming.state_decision_dataset_prepare.dataloader import TrajectoryDataset
import trajectory_data
from nocode_robot_programming.state_decision_dataset_prepare.dataset_auto import load_eval
loader = TrajectoryDataset(trajectory_data.package_path)
```

1. Datasets for all tasks. `loader.tasks.keys()`
2. Each task has dataset for each DS (decision state). `d_allDS`
3. Each DS dataset has `d_train, d_test, d_text`

We can see:
- Images in the datasets `showcase_captions`
- Images aligned at similar timesteps during demonstration with `showcase_aligned`.
- Play video for given task `loader.play_video('p0b_peg_pick')`.
- Plot task graph for given task `loader.plot_task_graph(skill_name="p0b_peg_pick")`.

Notes:
`p1b_peg_pick` - Long distance to the peg


# Generate dataset for all tasks

```python
datasets  = [load_eval(loader, name) for name in list(loader.tasks.keys())]
```

See \# of datasets

```python
len(datasets)
```

# Check all image samples for all datasets

```python
for d_allDS in datasets:
    for d_train, d_test, d_text in d_allDS:
        print(d_text)
        d_train.showcase_captions(scale=2.0)
        d_test.showcase_captions(scale=2.0)
```

# Extracted images at the similar timesteps aligned

```python
[d_train.showcase_aligned(scale=1.0)
    for d_allDS in datasets
    for d_train, d_test, d_text in d_allDS
]        
```
