```python
import torch; import numpy as np

def add_regulation_synthetic_images(d_train):
    add_size_half = d_train.X.shape[0] // 5
    add_size = add_size_half * 2
    
    X = torch.vstack((
        d_train.X,
        torch.zeros((add_size_half,d_train.X.shape[1],d_train.X.shape[2])).cuda(),
        torch.ones((add_size_half,d_train.X.shape[1],d_train.X.shape[2])).cuda()
    ))

    Xt = torch.hstack((
        d_train.Xt,
        torch.Tensor(np.random.choice(d_train.Xt.cpu(), size = add_size)).cuda(),
    ))

    if 'anomaly' not in d_train.y_cls:
        y_cls = d_train.y_cls + ["anomaly"]
        
        anomaly_int = d_train.y_int.max() + 1
    else:
        y_cls = d_train.y_cls
        anomaly_int = d_train.y_cls['anomaly']
    
    y_int = torch.hstack((
        d_train.y_int,
        torch.tensor(add_size * [anomaly_int]).cuda()
    ))
    y_names = d_train.y_names + add_size * ['anomaly']
    return X, Xt, y_cls, y_int, y_names
X, Xt, y_cls, y_int, y_names = add_regulation_synthetic_images(d_train)
```
