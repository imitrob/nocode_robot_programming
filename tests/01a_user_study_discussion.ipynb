# Notes

We want to keep the user study short so the entire time doesn't exceed 60 minutes.
From our 8 people, we had around a half experienced with some kinesthetic and joystick robot operation.
The others were novices. We can see the differences of demonstration execution accuracy for each person.
No person was experienced in gesture robot teleoperation.

The observed problems:
- The user forgot the correct command: Even if introduced, users confused a peg with the probe. Users did fast motions, even when they were told to keep the demonstration safe and slow. 
- Users would do better demonstrations with the second attempt, however we didn't want to tire the people. We allowed the second demonstration attempt on around 2% of demonstrations, where there was obvious user planning misunderstanding. 
- Some users wrapped from cable the more difficult way even when they were adviced of a better solution. This resulted in many demonstrations not being successful, because of cable slipping. During any task, there are usually more ways to solve it. The quality of demonstration is proportional to the user manipulation expertice.
- Our solution is reliable when the scene includes discrete state cases, where for each case, we have a solution. This is a problem for infinite states that the position of the probe can have. In this case, we don't want each state to have a separate demonstration, we want to have a continous method that solves each pose of the 


We observed the learning curve. Such as with the gesture teleoperation, where we observed user's gaining skills during the user study. Some users matched the precision of the gestures with the kinesthetic teaching and did fine motions, others wanted to finish the task as fast as possible and head quickly for the goal despite a robot moving fast and doing collisions.


We observed users taking different ways to solve the task even with these easy tasks. For example, some users branched early at opened door and opened door with the pulling aid attached to door. Others branched after picking the probe and opened a door by the probe itself.


Users should see what the robot will do, what part it will manipulate, if the action is reversible or not.

When the configuration of the robot was not usual, such as, end-effector far from the base, we observed an assigned force limits manifested as the precision of the target eef pose had an error and slide slowly away, we need to see how severe this problem is and make conditions, where it happens to mitigate it.
