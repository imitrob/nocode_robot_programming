# Artificial dataset generation

1. Use predefined dataset generator and loader with `d1_move`
2. Play with constructing your artificial dataset with `generate_run`

```python
from nocode_robot_programming.state_decision.dataset_D2 import d1_move

# Loading artificial dataset
d_train, d_test, d_text = d1_move(None)[0]
print(f"{d_text=},\n{d_train.X.shape=},\n{d_train.y_int.shape=},\n{len(d_train.y_names)=},\n{d_train.y_cls=}")
```

Here, we generate an artificial dataset with two labels: Taskboard with and without peg inserted.

We test by generating runs with offset.

```python
from nocode_robot_programming.artificial_dataset.dataset_generator import generate_run, RunArgs

generate_run(RunArgs(folder="d1_train", cls="base", offset=(0,0,0), mesh="taskboard.stl"))
generate_run(RunArgs(folder="d1_test",  cls="off" , offset=(8,0,0), mesh="taskboard.stl"))
generate_run(RunArgs(folder="d1_train", cls="base", offset=(0,0,0), mesh="taskboard_nopeg.stl"))
generate_run(RunArgs(folder="d1_test",  cls="off" , offset=(8,0,0), mesh="taskboard_nopeg.stl"))
```

`generate_run` generates dataset images at `trajectory_data/trajectories/artificial_dataset`

or visualize it:

```python
from nocode_robot_programming.state_decision.dataset_d2 import get_dataset_view_artificial_dataset
d_train = get_dataset_view_artificial_dataset(folder="d1_train")
d_test = get_dataset_view_artificial_dataset(folder="d1_test", y_cls=d_train.y_cls)

d_train.showcase_captions(fps=20, scale=2.0)
d_test.showcase_captions(fps=20, scale=2.0)
```
